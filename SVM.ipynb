## Introduction

Support Vector Machines (SVMs) are a type of supervised machine learning algorithm used for classification and regression tasks. They work by identifying the best possible boundary or **hyperplane** between classes. 

In binary problems, this is achieved by maximizinig the distance between data points of each class, and a given huperplane. This concept is called **maximum margin**. Data points that are closer to the hyperplane are called **support vectors** and influence the position and orientation of the hyperplane. The support vectors lie directly on and form the decision boundaries (also called bounding hyperplanes) which are parallel to the hyperplane and exist on each side of the hyperplane.

## Soft Margins

Soft margins help resole overlapping data by allowing data points to be in the margins. What this means is that the closest data points to a hyperplane might not necessarily be the support vectors that form the decision boundaries. Doing this often results in misclassification of those data points, but allow for a larger margin and thus better scalability in the long run when new data points are added.
